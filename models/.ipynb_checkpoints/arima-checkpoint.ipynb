# ARIMA

### Summary

ARIMA, short for auto-regressive integrated moving average, is a class of forecasting models based on the idea that the information in past values of a time series, alongside forecasting error and lag, can be used to predict the future values.

An ARIMA model is characterized by 3 terms: $p, d, q$, where:
* $p$ = order of the auto-regressive (AR) term
* $d$ = amount of differencing required to make the time series stationary
* $q$ = order of the moving average (MA) term

A pure auto-regressive (AR) model is one where $Y_{t}$ depends only on its own lag:

$Y_{t} = \alpha + \beta_{1}Y_{t-1} + \beta_{2}Y_{t-2} + \ldots + \beta_{p}Y_{t-p} + \epsilon_{t}$

A pure moving average (MA) model is one where $Y_{t}$ depends only on lagged forecast error:

$Y_{t} = \alpha + \phi_{1}\epsilon_{t-1} + \phi_{2}\epsilon_{t-2} + \ldots + \phi_{q}\epsilon_{t-q} + \epsilon_{t}$

An ARIMA model is a combination of both AR and MA equations, where the time series was differenced at least once to induce stationarity:

$Y_{t}^\prime = \alpha + \beta_{1}Y_{t-1}^\prime + \ldots + \beta_{p}Y_{t-p}^\prime + \phi_{1}\epsilon_{t-1} + \ldots + \phi_{q}\epsilon_{t-q} + \epsilon_{t}$

Some special cases of ARIMA models include:
* ARIMA(0, 0, 0): white noise
* ARIMA(0, 1, 0), without a constant: random walk
* ARIMA(0, 1, 0), with a constant: random walk with drift

### Caveats

Seasonal time series that do not exhibit patterns cannot be modeled with ARIMA models.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pmdarima as pm
from pmdarima.arima.utils import ndiffs
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
```

```python
# Read data from file.
df = pd.read_csv('./../data/raw/www_usage.csv', names=['value'], header=0)
df.head()
```

### Determining order of differencing.

The right order of differencing is the minimum differencing required to get a near-stationary time series to roam around a defined mean, where the auto-correlation function plot reaches zero fairly quickly.

If the autocorrelations are positive for a large number of lags, then the series needs further differencing. On the other hand, if the first lag's autocorrelation is too negative, then the series is probably over-differenced.

If the order of differencing is between two numbers, choose the order that results in the least standard deviation in the differenced series.

We'll also be examining the adjusted close data instead. As of the time of writing, Appleâ€™s stock has split five times since the company went public:

* 2-for-1 basis on June 16, 1987
* 2-for-1 basis on June 21, 2000
* 2-for-1 basis on February 28, 2005
* 7-for-1 basis on June 9, 2014
* 4-for-1 basis on August 28, 2020

```python
# Augmented Dickey-Fuller test.
result_adf = adfuller(df['value'])

print('ADF statistic: %f' % result_adf[0])
print('p-value: %f' % result_adf[1])
```

Since p-value < significance level, we won't need to difference the data at all. But for the sake of visualization:

```python
plt.rcParams.update({'figure.figsize':(9,7), 'figure.dpi':120})

# Original series.
fig, axes = plt.subplots(3, 2, sharex=True)
axes[0, 0].plot(df['value']); axes[0, 0].set_title('Original')
plot_acf(df['value'], ax=axes[0, 1])

# First order differencing.
axes[1, 0].plot(df['value'].diff()); axes[1, 0].set_title('First order differencing')
plot_acf(df['value'].diff().dropna(), ax=axes[1, 1])

# Second order differencing.
axes[2, 0].plot(df['value'].diff().diff()); axes[2, 0].set_title('Second order differencing')
plot_acf(df['value'].diff().diff().dropna(), ax=axes[2, 1])

plt.show()
```

For our data-set, it will appear that the time-series reaches stationarity with two orders of differencing, but the lag of the auto-correlation plot at two orders dips into the negatives fairly sharply, which indicates that the time-series may have been over-differenced.

Let's perform some tests:

```python
# Augmented Dickey-Fuller test.
result_adf = ndiffs(df['value'], test='adf')

# Kwiatkowski-Philips-Schmidt-Shin test.
result_kpss = ndiffs(df['value'], test='kpss')

# Philips-Perron test.
result_pp = ndiffs(df['value'], test='pp')

result_adf, result_kpss, result_pp
```

We'll use a differencing order of 1 to achieve weak stationarity.

### Determining order of auto-regression.

We will need to examine the partial autocorrelation function (PACF) plot, which displays the correlation between a time series and its lag, sans contributions from the intermediate lag.

Partial autocorrelation of lag $k$ of a times series is the coefficient of that lag in the autoregression equation of Y.

$Y_{t} = \alpha_{0} + \alpha_{1}Y_{t-1} + \alpha_{2}Y_{t-2} + \ldots + \alpha_{k}Y_{t-k}$

We will initially take the order of auto-regression to equal the lag that crosses the limit in the PACF plot.

```python
# Draw the partial autocorrelation function plot of the original series.
plt.rcParams.update({'figure.figsize':(9,3), 'figure.dpi':120})

fig, axes = plt.subplots(1, 2, sharex=True)
axes[0].plot(df['value'].diff()); axes[0].set_title('First order differencing')
axes[1].set(ylim=(0,0.5))
plot_pacf(df['value'].diff(), ax=axes[1])

plt.show()
```

### Determining order of moving average.

The autocorrelation function (ACF) plot can tell us the number of moving average terms we should use. The moving average can be thought of as the error of the lagged forecast.

```python
# Draw the autocorrelation function plot of the original series.
plt.rcParams.update({'figure.figsize':(9,3), 'figure.dpi':120})

fig, axes = plt.subplots(1, 2, sharex=True)
axes[0].plot(df['value'].diff()); axes[0].set_title('First order differencing')
axes[1].set(ylim=(0,1.2))
plot_acf(df['value'], ax=axes[1])

plt.show()
```

### Building the ARIMA model.

Now that we have defined $p$, $d$, and $q$, we have everything we need to build the ARIMA model.

```python
# ARIMA(1, 1, 2)
model = ARIMA(df['value'], order=(1,1,2))
model_fit = model.fit()

print(model_fit.summary())
```

```python
# ARIMA(1, 1, 1)
model = ARIMA(df['value'], order=(1,1,1))
model_fit = model.fit()

print(model_fit.summary())
```

The model's Akaike information criterion has reduced, and the p-values of the AR1 and MA1 terms have improved and are now much more statistically significant. We should plot the residuals and verify that no further patterns exist.

Notice here that the coefficient of the MA2 term is close to zero, and that the P-value in $P>|z|$ is highly insignificant. It should be less than 0.05 for the respective $X$ to be significant. We will rebuild the model with the MA2 term.

```python
# Plot residual errors.
residuals = pd.DataFrame(model_fit.resid)
fig, ax = plt.subplots(1,2)
residuals.plot(title="Residuals", ax=ax[0])
residuals.plot(kind='kde', title='Density', ax=ax[1])
plt.show()
```

The residual errors seem fine with near zero mean and uniform variance. We should be done here!*
