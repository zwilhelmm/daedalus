# Sanitize Data

Rearranges and re-indexes the data set to chronologically ascending order.

```python
import pandas as pd
```

```python
# Specify the filename.
filename = 'gld-adjusted-daily-full.csv'
```

```python
# Specify the read and write paths.
input='./../data/raw/{}'.format(filename)
output='./../data/clean/{}'.format(filename)
```

```python
# Read data from file.
df = pd.read_csv('{}'.format(input))
df.head()
```

```python
# Save the numerical index.
idx = df.index.values
```

```python
# Reverse the data set.
df = df.iloc[::-1]
df.head()
```

```python
# Reapplies the original index.
df.index = idx
df.head()
```

```python
# Check for any missing values.
clean = df.isnull().values.any()
```

```python
# Save to file.
df.to_csv('{}'.format(output), index=False)
```
